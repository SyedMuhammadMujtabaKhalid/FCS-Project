# 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
import io
import requests
import pandas as pd
import zstandard as zstd

base_url = "https://raw.githubusercontent.com/Shah-1510/Trending-YouTube-Dataset-FCS-2025-26-Project"

files = [
    "CAvideos.csv.zst", "DEvideos.csv.zst", "FRvideos.csv.zst", "GBvideos.csv.zst",
    "INvideos.csv.zst", "JPvideos.csv.zst", "KRvideos.csv.zst", "MXvideos.csv.zst",
    "RUvideos.csv.zst", "USvideos.csv.zst"
]

dfs = []
for file in files:
    country = file[:2]
    url = base_url + file
    print("Downloading:", url)
    resp = requests.get(url)
    resp.raise_for_status()

    dctx = zstd.ZstdDecompressor()
    # fully decompress into bytes
    with dctx.stream_reader(io.BytesIO(resp.content)) as reader:
        decompressed_bytes = reader.read()  # returns bytes

    # decode safely. errors='replace' will replace invalid sequences with ?
    # If you prefer to preserve bytes without decode errors, use latin1 instead:
    # text = decompressed_bytes.decode('latin1')
    text = decompressed_bytes.decode('utf-8', errors='replace')

    # read into pandas from a text buffer
    df = pd.read_csv(io.StringIO(text))
    df["country"] = country
    dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)
print("Combined rows:", len(final_df))
final_df.to_csv("final_df.csv", index=False)
```

# 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
final_df['publish_time'] = pd.to_datetime(final_df['publish_time'])

def interval_label(dt):
    start = dt.floor('10T')
    end = start + pd.Timedelta(minutes=10)
    return f"{start.strftime('%H:%M')} – {end.strftime('%H:%M')}"

final_df['publish_time_interval'] = final_df['publish_time'].apply(interval_label)
```

# 14. Read all json files with the video categories

```python
import pandas as pd
import json
import os

# Load JSON category files from local data/ folder
json_files = [
    "CA_category_id.json",
    "DE_category_id.json",
    "FR_category_id.json",
    "GB_category_id.json",
    "IN_category_id.json",
    "JP_category_id.json",
    "KR_category_id.json",
    "MX_category_id.json",
    "RU_category_id.json",
    "US_category_id.json"
]

all_categories = []

for file in json_files:
    country = file[:2]  # extract country code
    filepath = os.path.join("data", file)
    
    # Read JSON from local file
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    # Normalize items into dataframe
    df = pd.json_normalize(data["items"])
    
    # Add country column
    df["country"] = country
    
    all_categories.append(df)

# Combine all JSON files into one dataframe
categories_df = pd.concat(all_categories, ignore_index=True)

categories_df
```
